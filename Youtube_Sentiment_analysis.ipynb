```python
from textblob import TextBlob
```

```python
df.head()
```

```python
polarity =[]
for comments in df['comment_text']:
  try:
    polarity.append(TextBlob(comments).sentiment.polarity)
  except:
    polarity.append(0)
```

```python
 polarity
```

```python
df['polarity']  = polarity
```

```python
df.head()
```

```python
filter1 = df['polarity']==1
```

```python
positive_comments = df[filter1]
```

```python
filter2 = df['polarity']==-1
```

```python
Negative_comments = df[filter2]
```

```python
pip install wordcloud
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
set(STOPWORDS)
```

```python
total_comments_positive = ' '.join(positive_comments['comment_text'])
```

```python
wordcloud = WordCloud(stopwords=set(STOPWORDS)).generate(total_comments_positive)
```

```python
plt.imshow(wordcloud )
plt.axis('off')
```

```python
total_comments_negative = ' '.join(Negative_comments['comment_text'])
```

```python
wordcloud_neg = WordCloud(stopwords=set(STOPWORDS)).generate(total_comments_negative)
```

```python
plt.imshow(wordcloud_neg )
plt.axis('off')
```

```python
emoji_list = []
for comment in df['comment_text'].dropna():
  for char in comment:
    if char in emoji.EMOJI_DATA:
      emoji_list.append(char)

```

**bold text**

```python
emoji_list
```

```python
from collections import Counter
```

```python
Counter(emoji_list).most_common(10)
```

```python
emojis =[Counter(emoji_list).most_common(10)[i][0] for i in range(10)]
```

```python
freq = [Counter(emoji_list).most_common(10)[i][1] for i in range(10)]
```

```python
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

#init_notebook_mode(connected=True)
```
